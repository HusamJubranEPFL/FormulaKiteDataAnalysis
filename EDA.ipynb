```python
def detect_simple_sign_switch(df, value_col='TWA'):
    data = df[value_col].values
    signs = np.sign(data)

    switch_indices = np.where(signs[1:] != signs[:-1])[0] + 1  # +1 car comparaison (n vs n-1)

    # Ajout du premier et du dernier index si le DataFrame n'est pas vide
    if len(df) > 0:
        switch_indices = np.concatenate(([0], switch_indices, [len(df) - 1]))

    return df.loc[switch_indices, ['Lat', 'Lon', value_col, 'SecondsSince1970']]\
             .assign(index=switch_indices)

selected_maneuver = pd.DataFrame([
min([result_sb.iloc[i], result_kl.iloc[i]], key=lambda row: row['SecondsSince1970'])
for i in range(min(len(result_sb), len(result_kl)))
])

result_sb_man = detect_simple_sign_switch(senseboard_df)
result_kl_man = detect_simple_sign_switch(karl_df)
```

```python
def detect_TWA_change(df, value_col='TWA'):
    data = df[value_col].values
    value_diffs = np.abs(data[1:] - data[:-1])

    switch_indices = np.where(value_diffs > 10)[0] + 1  # +1 car comparaison (n vs n-1)

    return df.loc[switch_indices, ['Lat', 'Lon', value_col, 'SecondsSince1970']]\
             .assign(index=switch_indices)

result_sb = detect_TWA_change(senseboard_df)
result_kl = detect_TWA_change(karl_df)
```

```python
from itertools import combinations
from typing import Tuple

def find_best_ranges(df, index_col: str) -> Tuple[Tuple[int, int], Tuple[int, int]]:
    """
    Returns the two non-overlapping index ranges with the largest total span.
    """
    indices = sorted(df[index_col].tolist())
    ranges = list(combinations(indices, 2))

    best_pair = max(
        ((r1, r2) for r1, r2 in combinations(ranges, 2) if r1[1] < r2[0] or r2[1] < r1[0]),
        key=lambda pair: (pair[0][1] - pair[0][0]) + (pair[1][1] - pair[1][0]),
        default=((None, None), (None, None))
    )

    return best_pair

def trim_ranges_to_non_overlap(r1: Tuple[int, int], r2: Tuple[int, int]) -> Tuple[Tuple[int, int], Tuple[int, int]]:
    """
    Trim two potentially overlapping ranges to make them strictly non-overlapping.
    """
    a1, a2 = r1
    b1, b2 = r2

    if a2 <= b1 or b2 <= a1:
        # Already non-overlapping
        return r1, r2

    # Determine order
    if a1 <= b1:
        return (a1, min(a2, b1)), (max(a2, b1), b2)
    else:
        return (b1, min(b2, a1)), (max(b2, a1), a2)

# Apply to both datasets
r1_sb, r2_sb = find_best_ranges(result_sb_man, "index")
r1_kl, r2_kl = find_best_ranges(result_kl_man, "index")

print("Best ranges for result_sb_man:", r1_sb, "and", r2_sb)
print("Best ranges for result_kl_man:", r1_kl, "and", r2_kl)

# Trim corresponding ranges to remove overlaps
final_r1, final_r2 = trim_ranges_to_non_overlap(r1_kl, r2_kl)

print("Final trimmed ranges:", final_r1, "and", final_r2)
```

```python
from typing import Tuple
import pandas as pd

def max_successive_gap_in_range(df: pd.DataFrame, index_col: str, range_: Tuple[int, int]) -> Tuple[int, int, int]:
    """
    Compute the maximum gap between successive points in a given index range.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        index_col (str): Name of the column with index values.
        range_ (Tuple[int, int]): (start, end) of the index range to consider.

    Returns:
        Tuple[int, int, int]: (start_index, end_index, gap)
    """
    # Filter to the specified range
    mask = (df[index_col] >= range_[0]) & (df[index_col] <= range_[1])
    indices = sorted(df.loc[mask, index_col].tolist())

    max_gap = 0
    max_pair = (None, None)

    for i in range(1, len(indices)):
        gap = indices[i] - indices[i - 1]
        if gap > max_gap:
            max_gap = gap
            max_pair = (indices[i - 1], indices[i])

    return max_pair[0], max_pair[1], max_gap

# These are your final trimmed ranges from earlier logic
result_all = pd.concat([result_sb_man,result_sb, result_kl_man, result_kl], ignore_index=True)

# Apply to unified DataFrame
first_gap = max_successive_gap_in_range(result_all, "index", final_r1)
second_gap = max_successive_gap_in_range(result_all, "index", final_r2)

print("Max gap in first range:", first_gap[0], first_gap[1]  )
print("Max gap in second range:", second_gap[0], second_gap[1])
```
