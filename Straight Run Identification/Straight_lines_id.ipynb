```python
import os
from twa_analysis import analyze_session

# Dossier racine
base_dir = "../Data_Sailnjord/Straight_lines"

# Itère sur chaque dossier de date (ex: 06_06, 07_06, ...)
for date_folder in os.listdir(base_dir):
    date_path = os.path.join(base_dir, date_folder)
    if not os.path.isdir(date_path):
        continue  # ignore les fichiers

    # Itère sur chaque run dans le dossier date
    for run_folder in os.listdir(date_path):
        run_path = os.path.join(date_path, run_folder)
        if not os.path.isdir(run_path):
            continue

        # Liste les fichiers .csv dans le run
        csv_files = [f for f in os.listdir(run_path) if f.endswith(".csv")]
        if len(csv_files) != 2:
            print(f"⚠️ Skipping {run_path}, expected 2 CSV files, found {len(csv_files)}")
            continue

        # Obtenir les chemins complets
        csv_paths = [os.path.join(run_path, f) for f in csv_files]

        print(f"✅ Analyzing {csv_files[0]} and {csv_files[1]} from {run_path}")
        try:
            analyze_session(csv_paths[0], csv_paths[1])
        except Exception as e:
            print(f"❌ Error analyzing {run_path}: {e}")
```
